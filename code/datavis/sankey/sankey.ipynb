```python
import pandas as pd

data = pd.read_json(r'..\..\..\data\data_for_sankey.json')
```

```python
#学院情况
plantCountByCollege = {}

#植物类型
plantTemplateOfDonated = {}

for i in range(len(data)):
  #owner来判断是否已被认养
  if(data.loc[i,['owner']].isnull().values.any() == False):
    #学院
    college = data.loc[i]['owner']['title']
    if college not in plantCountByCollege:
        plantCountByCollege[college] = {
            'name': college,
            'target': 0,
        }
    plantCountByCollege[college]['target'] += 1

    #类型
    plantType = data.loc[i]['template']['name']
    if plantType not in plantTemplateOfDonated:
        plantTemplateOfDonated[plantType] = {
            'name': plantType,
            'target': 0,
        }
    plantTemplateOfDonated[plantType]['target'] += 1

# 对字典进行排序,将排序后的结果转换为字典
plantCountByCollege = dict(sorted(plantCountByCollege.items(), key=lambda x: x[1]['target'], reverse=True))
```

```python
mycolors =  [
  '#b3de69', '#8dd3c7', '#80b1d3', '#bebada', '#DC8FD9', '#fccde5', '#EB0D00', '#EB6034', '#EB9623', '#FFE65F',
  '#b3de69', '#8dd3c7', '#80b1d3', '#bebada', '#DC8FD9', '#fccde5', '#EB0D00', '#EB6034', '#EB9623', '#FFE65F',
  '#b3de69', '#8dd3c7', '#80b1d3', '#bebada', '#DC8FD9', '#fccde5', '#EB0D00', '#EB6034', '#EB9623', '#FFE65F']
c = 0
```

```python
# 节点
nodes = []
# 边
linkobjects = []

#总结点
nodes.append({'name': '全部认领植物','itemStyle':{'color':mycolors[c]}})
c += 1

#学院
for college in plantCountByCollege:
    node = {
      'name': plantCountByCollege[college]['name'],
      'itemStyle':{'color':mycolors[c]}
    }
    c += 1
    nodes.append(node)

    linkobject = {
      'source': plantCountByCollege[college]['name'],
      'target': '全部认领植物',
      'value': plantCountByCollege[college]['target'],
    }
    linkobjects.append(linkobject) 

#植物种类
# 植物
for plantType in plantTemplateOfDonated:
  node = {
    'name': plantTemplateOfDonated[plantType]['name'],
    'itemStyle':{'color': mycolors[c] }
  }
  c += 1
  nodes.append(node)

  linkobject = {
    'source': '全部认领植物',
    'target': plantTemplateOfDonated[plantType]['name'],
    'value': plantTemplateOfDonated[plantType]['target'],
  }
  linkobjects.append(linkobject) 
```

```python
from pyecharts import options as opts
from pyecharts.charts import Sankey


sankey = (Sankey()
    .add(
        "sankey",
        nodes, 
        linkobjects,
        layout_iterations = 0, #按原顺序
        node_gap = 10, #同一列之间结点的间隔
        node_width = 30, #结点矩形的宽度
        focus_node_mode = 'adjacency',
        pos_left = "25%", #容器距离左侧
        pos_right = "5%", #容器距离右侧
        pos_top = "8%", #容器距离上侧
        linestyle_opt = opts.LineStyleOpts(opacity = 0.45, curve = 0.75, color = "source"),
        label_opts = opts.LabelOpts(position = "left", font_size  = 12, distance = 12),
        itemstyle_opts = opts.ItemStyleOpts(border_width = 1.5, border_color = "#aaa"),
        tooltip_opts = opts.TooltipOpts(trigger_on = "mousemove"),
    )
    
    .set_global_opts(title_opts=opts.TitleOpts(title = "认领植物数量")))

sankey.load_javascript()
sankey.render('sankey.html')

sankey.render_notebook()
```
